```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

annual_df = pd.read_excel('Table_1.1_Primary_Energy_Overview.xlsx', skiprows=1)

annual_df = annual_df.drop(0)

annual_df.columns = [
    'Year',
    'Total Fossil Fuels Production',
    'Nuclear Electric Power Production',
    'Total Renewable Energy Production',
    'Total Primary Energy Production',
    'Primary Energy Imports',
    'Primary Energy Exports',
    'Primary Energy Net Imports',
    'Primary Energy Stock Change and Other',
    'Total Fossil Fuels Consumption',
    'Nuclear Electric Power Consumption',
    'Total Renewable Energy Consumption',
    'Total Primary Energy Consumption'
]

annual_df['Year'] = annual_df['Year'].astype(int)

plt.figure(figsize=(12, 6))
plt.plot(annual_df['Year'], annual_df['Total Primary Energy Production'], marker='o', linestyle='-')
plt.title('Total Primary Energy Production Over the Years')
plt.xlabel('Year')
plt.ylabel('Energy Production (Quadrillion Btu)')
plt.grid(True)

ax = plt.gca()
ax.xaxis.set_major_locator(mticker.MultipleLocator(10))
plt.xticks(rotation=45) # Rotate labels for better readability

plt.tight_layout()

plt.savefig('primary_energy_production_with_year_ticks.png')

plt.show()
```
